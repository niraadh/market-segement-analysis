```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
behavior = pd.read_csv('purchase_behaviour.csv')
transactions = pd.read_csv('transaction_data.csv')
```

```python
behavior.head()
```

```python
transactions.head()
```

```python
merged_df = pd.merge(transactions, behavior, on='LYLTY_CARD_NBR', how='left')
merged_df.head()
```

```python
customer_segment_sales = merged_df.groupby(['LIFESTAGE', 'PREMIUM_CUSTOMER'])['TOT_SALES'].sum().reset_index()
customer_segment_sales.sort_values(by='TOT_SALES', ascending=False)
```

```python
avg_segment_sales = merged_df.groupby(['LIFESTAGE', 'PREMIUM_CUSTOMER'])['TOT_SALES'].mean().reset_index()
avg_segment_sales.sort_values(by='TOT_SALES', ascending=False)
```

```python
customer_segment_sales = merged_df.groupby(['LIFESTAGE', 'PREMIUM_CUSTOMER'])['TOT_SALES'].count().reset_index()
customer_segment_sales.sort_values(by='TOT_SALES', ascending=False)
```

```python
#the above tables show
#(1)conrtibution of each segment towards the total sales
#(2)each segments average value
#(3)segment with most frequent costumers
```

```python
product_sales = merged_df.groupby('PROD_NAME')['TOT_SALES'].sum().reset_index()
top_products = product_sales.sort_values(by='TOT_SALES', ascending=False)
top_products
```

```python
#shows the top 3 best selling products to be 
#Dorito Corn Chp Supreme 380g
#Smiths Crnkle Chip Orgnl Big Bag 380g
#Smiths Crinkle Chips Salt & Vinegar 330g
```

```python
customer_total_sales = merged_df.groupby(['LYLTY_CARD_NBR','LIFESTAGE', 'PREMIUM_CUSTOMER'])['TOT_SALES'].sum().reset_index()
loyal_customers = customer_total_sales.sort_values(by='TOT_SALES', ascending=False)
loyal_customers.head(20)
```

```python
#most loyal customers are majorly families either young or old
#most loyal cotermer majorly prefer premium or buget products
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='TOT_SALES', y='LIFESTAGE', data=customer_segment_sales)
plt.title('Total Sales by Customer Segment (Lifestage)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='TOT_SALES', y='LIFESTAGE', hue='PREMIUM_CUSTOMER', data=customer_segment_sales)
plt.title('Total Sales by Customer Segment (Lifestage & Premium Status)')
plt.show()
```
